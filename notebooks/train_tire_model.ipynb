# Train Tire Model
Notebook placeholder for EDA and training

```python
# Install required packages
# !pip install fastf1 scikit-learn pandas

import fastf1
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import joblib

# Enable fastf1 caching to speed up data loading
fastf1.Cache.enable_cache('./f1_cache') 

# Load a specific race (e.g., 2025 British Grand Prix)
session = fastf1.get_session(2025, 'Silverstone', 'R')
session.load(laps=True, weather=True)

# Process lap details into dataframes
laps = session.laps
laps['LapTime_Sec'] = laps['LapTime'].dt.total_seconds()

# Feature Engineering: Clean data and track features
# We want to predict lap time based on Tire Compound, Tyre Life, and Track Temp
df = laps[['Driver', 'LapNumber', 'Stint', 'Compound', 'TyreLife', 'LapTime_Sec']].dropna()

# One-hot encode categorical tire compounds (SOFT, MEDIUM, HARD)
df = pd.get_dummies(df, columns=['Compound'])

# Define Features (X) and Target (y)
features = [col for col in df.columns if col not in ['Driver', 'LapTime_Sec']]
X = df[features]
y = df['LapTime_Sec']

# Train our Classical Supervised Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# Save the trained predictive engine
joblib.dump(model, '../ml_models/tire_deg_model.pkl')
joblib.dump(features, '../ml_models/model_features.pkl')
print("Model trained and saved successfully!")
```
